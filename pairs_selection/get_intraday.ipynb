```python
#url query inputs 
function = "TIME_SERIES_INTRADAY_EXTENDED"
ticker_lst = ["ES", "WEC"]
interval = "60min"
slice = [[f"year{j}month{i}" for i in range(1, 13)] for j in range(1, 3)]
slice = slice[0] + slice[1]
output = "json"
adjusted = True
urls = wrap.intraday_url(settings, function, ticker_lst, interval, slice, output = output, adjusted = adjusted)
```

```python
ES_data = threading.thrd_ticker_slices(urls["ES"])
WEC_data = threading.thrd_ticker_slices(urls["WEC"])
```

```python
ES_df = pd.DataFrame(data = ES_data, columns = ["time", "open", "high", "low", "close", "volume"])
WEC_df = pd.DataFrame(data = WEC_data, columns = ["time", "open", "high", "low", "close", "volume"])

ES_df["time"] = pd.to_datetime(ES_df["time"])
ES_df = ES_df.sort_values(by = ["time"])
WEC_df["time"] = pd.to_datetime(WEC_df["time"])
WEC_df = WEC_df.sort_values(by = ["time"])

display(ES_df)
display(WEC_df)
```
